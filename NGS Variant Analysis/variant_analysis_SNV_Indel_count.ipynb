```python
vcf_file = open('trio_example.vcf', 'r')
chrom_length = open('chromosome_lengths.tsv', 'r')
```

```python
total_mutations = 0
nu_mutations_N1 = 0
nu_mutations_N2 = 0
nu_mutations_N3 = 0

total_SNV = 0
nu_SNV_N1 = 0
nu_SNV_N2 = 0
nu_SNV_N3 = 0
for line in vcf_file:
    
    # check if the line is an info line, starting with ##, in that case, skip
    if (line.startswith("##")):
        continue
    if (line.startswith("#")):
        #here we want to find the index of each column
        header=line.split("\t") # split by tabulator
        print ("found the header" + str(header))
        for item in enumerate(header):
            if (item[1] == 'N1'):
                index_N1=item[0]
            if (item[1] == 'N2'):
                index_N2=item[0]
            if (item[1] == 'N3\n'):
                index_N3=item[0]
                
    #counting how many mutations we have in the file,then we break up the results for each individual          
    else:
        info = line.split("\t")
        N1=info[index_N1]
        N2=info[index_N2]
        N3=info[index_N3].strip()  # we use strip function to get rid of "\n" at the end of each line.
        
        
        total_mutations = total_mutations + 1
        
        if (N1 != "0/0"):
            nu_mutations_N1 = nu_mutations_N1 + 1
            
        
        if (N2 != "0/0"):
            nu_mutations_N2 = nu_mutations_N2 + 1
            
       
        if (N3 != "0/0"):
            nu_mutations_N3 = nu_mutations_N3 + 1
            
        #now we count how many SNVs and Indels for each indiviual    
        
        if (len(info[3])!=1):
            continue
        if (len(info[4])!=1):
            continue
            
        total_SNV = total_SNV + 1
        
        
        if (N1 != "0/0"):
            nu_SNV_N1 = nu_SNV_N1 + 1
        if (N2 != "0/0"):
            nu_SNV_N2 = nu_SNV_N2 + 1
        if (N3 != "0/0"):
            nu_SNV_N3 = nu_SNV_N3 + 1
            
           
        
        
        
print("total mutations : " + str(total_mutations)) 
print("number of mutations in N1 : " + str(nu_mutations_N1))
print("number of mutations in N2 : " + str(nu_mutations_N2))
print("number of mutations in N3 : " + str(nu_mutations_N3))

print ("~")
print("total SNVs : " + str(total_SNV))
print("SNVs in N1 : " + str(nu_SNV_N1))
print("SNVs in N2 : " + str(nu_SNV_N2))
print("SNVs in N3 : " + str(nu_SNV_N3))
    
print ("~")
total_Indels = total_mutations - total_SNV 
N1_Indels = nu_mutations_N1 - nu_SNV_N1
N2_Indels = nu_mutations_N2 - nu_SNV_N2
N3_Indels = nu_mutations_N3 - nu_SNV_N3

print ("total number of Indels : " + str(total_Indels))
print("Indels in N1 : " + str(N1_Indels)) 
print("Indels in N2 : " + str(N2_Indels))  
print("Indels in N3 : " + str(N3_Indels))  
```
